```python
from bokeh.plotting import figure, output_file, show, output_notebook
import pandas as pd
temp = pd.read_csv("FB_insights.csv", encoding = "ISO-8859-1")
```

```python
fbData = temp[1:len(temp.index)]
```

```python
fbData = fbData.apply(pd.to_numeric, args=('coerce', ))
```

```python
Anzahl_Fans_col = "Lifetime Total likes"
Anzahl_Fans_agg = fbData.iloc[len(fbData[Anzahl_Fans_col])-1][Anzahl_Fans_col]
Anzahl_Fans_line = list(fbData[Anzahl_Fans_col])
```

```python
Bezahlte_Reichweite_col = "Daily Paid reach"
Bezahlte_Reichweite_agg = fbData[Bezahlte_Reichweite_col].sum(axis=0)
Bezahlte_Reichweite_line = list(fbData[Bezahlte_Reichweite_col])
```

```python
Reichweite_col = "Daily Total reach"
Reichweite_agg = fbData[Reichweite_col].sum(axis=0)
Reichweite_line = list(fbData[Reichweite_col])
```

```python
Organische_Reichweite_agg = Reichweite_agg - Bezahlte_Reichweite_agg
Organische_Reichweite_line = list(fbData[Reichweite_col] - fbData[Bezahlte_Reichweite_col])
```

```python
Organische_Reichweite_Pct_agg = Organische_Reichweite_agg / Reichweite_agg
# Organische_Reichweite_Pct = "{:.1%}".format(Organische_Reichweite_Pct)
Organische_Reichweite_Pct_line = list(Organische_Reichweite_line / fbData[Reichweite_col])
```

```python
Impressionen_col = "Daily Total impressions"
Impressionen_agg = fbData[Impressionen_col].sum(axis=0)
Impressionen_line = list(fbData[Impressionen_col])
```

```python
PTA_col= "Daily People talking about this"
PTA_agg = fbData[PTA_col].sum(axis=0)
PTA_line = list(fbData[PTA_col])
```

```python
# for the plotting
days = list(range(1,len(fbData)+1))
# ToDo: Dates statt einfach nur Zahlen

var1 = Anzahl_Fans_line

output_notebook(hide_banner=True)

# for the Fans

p = figure(
   tools="pan,box_zoom,reset,save", title="Anzahl Fans",
   x_axis_label='Die letzten 28 Tage', y_axis_label='Lifetime Total likes'
)

# add some renderers
p.line(days, var1, legend="Fans", line_width=1, line_color="green", line_dash = "4 4")
p.circle(days, var1, legend="Fans", fill_color="green", size=8)

# show the results
show(p)
```

```python
# for the plotting
days = list(range(1,len(fbData)+1))
# ToDo: Dates statt einfach nur Zahlen

var1 = Reichweite_line
var2 = Organische_Reichweite_line
var3 = Bezahlte_Reichweite_line


output_notebook(hide_banner=True)

# for the Fans

p = figure(
   tools="pan,box_zoom,reset,save", title="Reichweite",
   x_axis_label='Die letzten 28 Tage', y_axis_label='Reichweite'
)

# add some renderers
p.line(days, var1, legend="Reichweite", line_width=1, line_color="blue")
p.circle(days, var1, legend="Reichweite", fill_color="blue", size=8)
p.line(days, var2, legend="Organische Reichweite", line_width=1, line_color="green", line_dash = "4 4")
p.line(days, var3, legend="Bezahlte Reichweite", line_width=1, line_color="red", line_dash = "4 4")

# show the results
show(p)
```
